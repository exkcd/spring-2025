## How do I pick colors?

### Color temperature

Cool vs. warm

Cool = blue, indigo, purple

Warm = red, orange, yellow

Green is weird...

* can be cool or warm

### Analogous colors

Colors adjacent to one another on the color wheel

E.g. red, orange, yellow

### Complementary colors

Colors across from each other on the color wheel

#### Split complementary

Pick a color, find it's complement, and then choose the analogous colors beside it

### Accent colors

A color that stands out in the palette

* Can be achieved by taking a color and whitening it

### Monochrome

A "one color" color palette

* Tinting and shading a singular color

## How do I pick fonts?

### Typeface vs. font

* Typeface: What we call a "font" (i.e. Arial, Helvetica, Courier New)
* Font: Thin, Light, Regular, Bold, Extra Bold
    * Variation of a typeface

### Serif vs. sans serif

* Serifs come from the printer press
    * Increase legibility
    * Used for body text
* Sans serif usually used for headings and not a lot of reading

### Spacing

Used for:

* Tracking
    * Letter spacing
* Kerning
    * Space between any two adjacent letters
    * Why use? Sometimes used to adjust space between individual letters
    * Quite important in graphic design
* Leading
    * Space between lines
    * "Line height"
    * Space between individual lines
    * Why use tight lines? Save money with printing :))
    * Comes from printing press
    * Needs to be legible
