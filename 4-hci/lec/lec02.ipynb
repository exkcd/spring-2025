## Recap

tint = lighter version of a color
shade = darker version of a color

### Complementary colors

* Colors on the opposite ends of the color wheel

#### Split complementary

* Using two colors to either side of the complement giving you a three-color color palette

### Analogous colors

* Color palette that uses colors next to each other on the color wheel

### Monochromatic

* Color palette using only tints and shades of one color

### Skeuomorphism

* Making images look like their real world counterpart

### Patterns

* Seamless
    * Meant to tile in every direction

* Semi-seamless
    * Tiles in one direction (e.g. horizontal), but not the other (e.g. vertical)

* Non-seamless
    * Can't be tiled in any direction
    * Can only be used once on a webpage

### Aspect ratio
