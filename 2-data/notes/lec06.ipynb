## EDA and Visualization

### Visualizing Quantitative Features

* Bar plots aren't the *best* for quantitative data

Good plots:

* Box plot
* Violin plot
* Histogram

#### Histogram

* Density is a good measure of data points in histogram
    * Proportion of data points per unit of x-axis

#### Counting bin size and density

* Find the proportion in a bin
    * Given width of a bin (x), and height of a bar (y)
    * Proportion = x * y
    * Total data points * Proportion

#### Density Curves

* Kernel Density Estimate (density curve)
    * Used to estimate a probability density function
* Provides a continuous distribution to discrete data points
* Area underneath a curve can help us understand probability

#### Bar Chart vs Histogram

Bar chart:

* Distribution of ONE qualitative variable
* Bars have arbitrary (but equal) widths and spacings
* heigh of bars proportional to the percent of individuals

Histogram

* Distribution of one quantitative variable
* Horizontal axis is numerical
* Density histogram: area of bars = proportion of data points in a bin
* height measures density

#### Percentiles

The *n*th percentile is that value *q* such that *n\%* of the data values fall above or below it

* 'q' might not be unique
* should be little difference between these definitions
* 'cutoffs' for areas

* \[first quartile, third quartile\] contains the "middle 50%" of the data

##### Inter-quartile range

* IQR = third quart - first quart

### Box-plot

* width = meaningless
* box = lower, middle, upper quartile
* whiskers
    * largest data point within 1.5*IQR from upper quartile
    * smallest data point within 1.5*IQR from lower quartile
    * outside = outliers

THIS IS CONVENTION (ARBITRARY DEFINITIONS)
(this is how we will draw plots in this class)

### Violin plot

* similar to box plots, but also show smoothed density curves
* width = density
