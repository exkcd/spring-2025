## Variables

### Variables are columns

What does each column represent?
* Variable is a measurement of a particular concept

2 common properties
* datatype/storage type
    * how each variable value is stored in memory
    * int, float, bool, etc.
* variable type/feature type
* storage types say what operations we can code
conceptualized measurement of info
    * use expert knowledge

### Variable feature types

Not all fit neatly into these two categories!

* quantitative
    * ratios and intervals have consistent meaning
    * continuous
        * could be measured ot arbitrary precision
            * price
            * temperature
    * discrete
        * stored as ints where intervals have consistent meaning
            * \# of siblings
            * years of education
* qualitative
    * ordinal
        * categories w/ ordered levels; no consistent meaning ot difference
            * preferences
            * level of education
    * nominal
        * categories w/ no specific ordering
            * political affiliation
            * CU ID number

## Faithfulness

Does my data contain unrealistic or incorrect values?

* Dates in the future for events in the past
* Locations that don't exist
* Negative counts
* Misspellings of names
* Large outliers

### Signs that your data may not be faithful

* Truncated data
    * Early Microsoft Excel limited data to 65536 rows and 255 cols
* Spelling errors
* Time zone inconsistencies
* Duplicated records/fields
* Units not specified

### How to handle missing data

* Drop records with missing values
    * Only drop on training data
    * Don't drop on testing/validation sets
    * Missing or corrupt records might be related to something interesting
* Keep as NaN
* Imputation/Interpolation
    * Infer from missing values
    * Average imputation: replace with the mean
    * Median imputation: replace with the median
    * Mode imputation: replace with the mode
    * Zero: replace w/ 0 if it makes sense in context
